```python
import random
word=''
pane=[]  #to display the unguessed state of word
f=open('words.txt','r')

secret=[]
for l in f:
    l=l.strip('\n')
    secret.append(l)
word=random.choice(secret)
pane=['?']*len(word)   #if word is 'goal'  then pane=['?','?','?','?']
    
```

```python
def lifePane(lives):                    #printing the life count display
    for i in range(1,10):
        if i<=lives:
            print('\u2661',end='') #unicode for heart symbol
           
        else:    
            print('\u2620',end='')#unicode for skull and crossbones
    print(':'+str(lives))
    
def setLives(lives):                    #UPDATING LIVES COUNT AND PRINTING MESSAGE
    lives-=1
    
    print("Sorry!Invalid guess!")
    return lives
def winnerMessage():                   #PRINTING WINNER MESSAGE
    print("Hurrah! Well done!")
```

```python
# MAIN GAME CODE
lives=9
guesses=0
correct=False

while(lives>0 and not correct):
    
    print('**************************************************************')
    lifePane(lives)
    print(pane,end='\n')
    g=input("Enter a letter or a whole word:")
   
    if len(g)!=1 and len(g)!=len(word):                 #FILTER to accept only letter or the whole word of same length
        lives=setLives(lives)
    else:
        if len(g)==len(word) and lives>0:               # the full word entry branch
            if g.lower()==word:
                correct=True
            else:
                lives=setLives(lives)
        elif len(g)==1 and lives>0:                    # the single letter input
            if g in word:              # if letter in word, update '?' pane                
                for i in range(len(word)):
                    if word[i]==g:
                        pane[i]=g
            else:                     # if letter not in word,update lives count, decrease by 1
                lives=setLives(lives)
        if '?' not in pane: # if all letters correctly guessed!
            correct=True
            print(pane)
if(correct and lives>0): # checking for winner
    print('**************************************************************')
    winnerMessage()
else:
    print('Game Over! \u2620')
    print('the correct word was:{}'.format(word))
word=''
pane=['?']*5
```
