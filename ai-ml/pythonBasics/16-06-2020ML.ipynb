```python
import numpy as np
import pandas as pd
import sklearn as sk
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/weather_data.csv",parse_dates=["day"])
data
```

```python
data=data.replace({"temperature":-99999,"windspeed":-99999,"event":'0'},np.nan)
data
```

## interpolate:


```python
int_data=data.interpolate()
int_data
```

```python
##dropna:
drop_data=data.dropna(how='any',axis=1)
drop_data
```

```python
drop_all=data.dropna(how='all')
drop_all
```

<font color="green"><h1 align="center"> Handling Missing Values using Scikit Learn</h1>

## simpleImputer

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
```

```python
imputer.fit(data.iloc[:,1:3]) 
```

```python
data.iloc[:,1:3]=imputer.transform(data.iloc[:,1:3])
data
```
