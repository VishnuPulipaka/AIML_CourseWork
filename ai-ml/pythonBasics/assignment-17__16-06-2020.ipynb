## Read the dataset from the below URL in a proper format:
https://data.princeton.edu/wws509/datasets/salary.dat

```python
df=pd.read_table("https://data.princeton.edu/wws509/datasets/salary.dat",sep="\s+")    # using separator as Regular expression
                                                                                       # "\s+" one or more spaces
df
```

## 1. Check for any missing values in the dataset

```python
df[df.isnull().values]
```

### conclusion : no missing values

## 2. Display the data of only the numerical columns

```python
numerics=df.select_dtypes(include=np.number)
numerics
```

## 3.Display the data of only the non-numerical columns

```python
nonNum=df.select_dtypes(exclude=np.number)
nonNum
```

## 4.Get a list of categorical variables with the categories frequency count.

```python
for i in nonNum.columns:
    
    print(nonNum[i].value_counts(),"\n")
```

## 6. Create a dataframe by adding dummy variables to the original data frame using concat.
## 9. Also, drop one of the dummy columns created for sx and dg variables

```python
dummy_set=pd.get_dummies(df.sx) 
dummy_set=dummy_set.drop(columns=['male'])        # dropping one of the columns!
dummy_set# dummy columns
```

```python
merged_data=df.copy()
```

```python
merged_data=pd.concat([merged_data,dummy_set],axis=1,sort=False)
```

```python
dummy_set=pd.get_dummies(df.dg)
dummy_set=dummy_set.drop(columns=['masters'])
dummy_set
```

```python
merged_data=pd.concat([merged_data,dummy_set],axis=1,sort=False)
merged_data
```

## 7. Create another  dataframe by adding dummy variables to the original data frame using merge
## 8.After adding the dummy columns to the dataframe, drop the original columns sx and dg 
## 9. Also, drop one of the dummy columns created for sx and dg variables

```python
ds1=pd.get_dummies(df.dg).reset_index()
ds2=pd.get_dummies(df.sx).reset_index()
```

```python
ds2=ds2.drop(columns=['male'])
ds1=ds1.drop(columns=['masters'])
```

```python
newd=df.copy().reset_index()
newd=newd.merge(ds2,how="inner",on='index')
newd=newd.merge(ds1,how="inner",on='index')
newd=newd.drop(columns=['index'])
newd
```

## 8. After adding the dummy columns to the dataframe, drop the original columns sx and dg 

```python
newd=newd.drop(columns=['sx','dg'])
```
