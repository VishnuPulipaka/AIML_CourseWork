```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/spam.csv",encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.v2,df.v1,test_size = 0.2,random_state = 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
wcv=cv.fit(X_train)
```

```python
len(wcv.get_feature_names())
```

```python
# transforming the data:
d_arr=wcv.transform(X_train).toarray()
d_arr
```

```python
pd.DataFrame(d_arr)
```

```python
feature_matrix=pd.DataFrame(d_arr,columns=wcv.get_feature_names())
feature_matrix
```

```python
# TFIDF Vectorizer

# Import TfIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

#Initialize and object for the TFIDF Vectorizer
tfidf= TfidfVectorizer()
```

```python
X_train_transformed=tfidf.fit_transform(X_train)
X_train_transformed
```

```python
X_test_transformed=tfidf.fit_transform(X_test)
X_test_transformed
```

```python
tfidf.get_feature_names()
```

#### Vocabulary:

```python
# position of the words in the sparse matrix
tfidf.vocabulary_
```

```python
# Idf of the terms
tfidf.idf_
```

```python
from sklearn.naive_bayes import BernoulliNB

model_BernNB = BernoulliNB()
```

```python
model_BernNB.fit(X_train_transformed,y_train)
```

```python
y_test_pred=model_BernNB.predict(X_test_transformed)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,y_test_pred)
```
