# Download the dataset from the below path and read it from your local machine using with Absolute Path or Relative Path

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading data using relative path:
df=pd.read_csv("company_sales_data.csv")
df.head()
```

# 1.Get the basic description from the data


```python
df.describe()
```

# 2. Get total profit of all months and show line plot with the following style properties. Generated line plot must include following style properties: –
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = month number
- Y label name = sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3


```python
plt.xlabel("month number")
plt.ylabel("sold units")
plt.title("Profit by month")
plt.plot(df["month_number"],df["total_units"],'-o',color='red',linestyle='dotted',lw='3')
plt.legend({"Total_units"},loc="lower right")


#saving the plot to hard disk to "plots" folder
plt.savefig('plots\\monthly_profit.png')
plt.show()
```

# 3. Read all product sales data and show it using a multi line plot

```python
# getting list of product names to check
products=df.columns.to_list()[1:7]
products
```

```python
plt.style.available
```

```python
plt.figure(figsize=(6,6))
plt.xlabel("Month number")
plt.ylabel("Units Sold")
# using product list to plot lines 
for i in products:
    plt.plot(df["month_number"],df[i],"-o",label=i,lw=4)
plt.grid()
plt.title("Product-wise Sales")
plt.legend(loc="best")

#saving plot to "plots" folder
plt.savefig("plots\\product_wise_sale.png")
```

# 4. Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(df.month_number,df.toothpaste,color="blue")
plt.title("Toothpaste Sales")
plt.xlabel("Month")
plt.ylabel("toothpaste sold(units)")
plt.savefig("plots\\toothpaste.png")
```

# 5. Read face cream and face wash product sales data and show it using the bar chart


```python
plt.figure(figsize=(15,6))
plt.title("Face products")
plt.xlabel("Month number")
plt.ylabel("Units Sold")
plt.bar(df.month_number-0.2,df.facecream,label="facecream",width=0.4)
plt.bar(df.month_number+0.2,df.facewash,label="facewash",width=0.4)
plt.legend(loc="best")
plt.savefig("plots\\faceproducts.png")
```

# 6. Read Bathing soap, face wash of all months and display it using the Subplot


```python

fig,a=plt.subplots(2)

a[0].plot(df.month_number,df[products[1]])
a[0].set_title("Face Wash")
a[1].plot(df.month_number,df[products[3]])
a[1].set_title("Bathing Soap")

#padding between subplots
fig.tight_layout(pad=3.0)
plt.savefig("plots\\soap.png")
plt.show()
```

# 7. Read the total profit of each month and show it using the histogram to see most common profit ranges

```python
plt.figure(figsize=(15,6))
plt.hist(df.total_profit,bins=[50000,100000,150000,200000,250000,300000,350000,400000,450000,500000])
plt.title("total monthly profit")
plt.xlabel("Profit value")
plt.ylabel("no of months")
plt.grid()
plt.savefig("plots\profitRanges.png")
plt.show()
```
