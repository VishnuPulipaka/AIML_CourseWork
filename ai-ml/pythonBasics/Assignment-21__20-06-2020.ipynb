# Convert the data type of integer cols to float and float cols to integer


```python
F2I=df.select_dtypes(include="float64").columns.to_list()
I2F=df.select_dtypes(include="int64").columns.to_list()
df[F2I]=df[F2I].astype('int64')
df[I2F]=df[I2F].astype('float64')
df.dtypes
```

# Get the row which has highest population

```python
df[df.population_total == df.population_total.max()]
```

# Calculate the mean of the population total grouped by States and sort the population 

```python
state_pop=df.groupby(df.state_name).mean().population_total
state_pop=pd.DataFrame(state_pop.sort_values(ascending=True))
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x=state_pop.population_total,y=state_pop.index)
```

# Create a dataframe of five cities which have lowest population


```python
low = df.nsmallest(5,'population_total')
low
```

# Create a dataframe of  five cities which have highest  population

```python
top = df.nlargest(5,'population_total')
top
```

# Concatenate the top five and bottom five cities based on the population

```python
low.sort_values('population_total',ascending = False,inplace = True)
new_df = pd.concat([top,low])
new_df

```
