```python
## OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Insider the columntransformer,we have to pass the OneHotEncoder
column_transformer=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],
                                     remainder='passthrough')
```

```python
df=column_transformer.fit_transform(df)
df
```

```python
X=pd.DataFrame(df)
X
```

# Splitting the Dataset

```python
# town and area -->input columns
price=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/homeprices.csv")
X=price.iloc[:,0:2]     # preparing the input variables/columns
X
y=price.iloc[:,2]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

# Feature Scaling: (Not mandatory for all the algorithms)

```python
from sklearn.preprocessing import scale

sample=pd.DataFrame({'W.E':[2,4,5,6,7,8],'Age':[28,25,32,45,33,32],
                   'Salary':[300000,400000,500000,600000,700000,800000],
                   'Distance':[10,12,23,32,15,17],
                   'Expected Salary':[400000,500000,600000,700000,800000,900000]})
sample
```

```python
sd=scale(sample)
sd=pd.DataFrame(sd)
sd
```

```python
## Using StandardScaler

from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
scd=SS.fit_transform(sample)
pd.DataFrame(scd)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
ms.fit(sample)
sample=ms.transform(sample)
pd.DataFrame(sample)
```
