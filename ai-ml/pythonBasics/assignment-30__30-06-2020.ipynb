# Apply the KNN Algorithm to the dataset without applying scaling operation and try to compare the results before scaling and after scaling.

```python
y=df.diagnosis
y.head()
X=df.drop(['id','diagnosis'],axis=1)
X
```

## Without Applying Scaling Operation

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30,random_state = 2)
```

```python
# Without Scaling the dataset
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6,metric = 'euclidean')
knn.fit(X_train,y_train)
```

```python
y_train_pred = knn.predict(X_train)
y_train_pred
```

```python
# Checking classification report
from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_pred))
```

```python
from sklearn.metrics import accuracy_score
score = []
for k in range(1,20):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train)
    pred_test = knn_model.predict(X_test)
    
    score.append(accuracy_score(y_test,pred_test))
print(score)
```

```python
# Plotting k values and scores
plt.plot(range(1,20),score,'--',marker = 'o',markerfacecolor = 'red')
```

#### Optimum K value is 9

```python
# Optimum k value is 9
final_model = KNeighborsClassifier(n_neighbors=9,metric = 'euclidean')
final_model.fit(X_train,y_train)
```

```python
# Predicting Train data
final_train_pred = final_model.predict(X_train)
```

```python
# Classification report
print(classification_report(y_train,final_train_pred))
```

```python
# Prediction on test data
final_test_pred = final_model.predict(X_test)
```

```python
# Classification report
print(classification_report(y_test,final_test_pred))
```

## With Scaling Operation

```python
# Preparing training and testing data
from sklearn.model_selection import train_test_split
X1_train,X1_test,y1_train,y1_test = train_test_split(X,y,test_size = 0.30,random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Scaling for train data
scaled_X1_train = pd.DataFrame(scaler.fit_transform(X1_train),columns = X1_train.columns)
# Scaling for test data
scaled_X1_test = pd.DataFrame(scaler.transform(X1_test),columns = X1_test.columns)
```

```python
## Fitting scaled model on knn and generating list of accuracy scores 
# Cheking for optimum k-value
from sklearn.metrics import accuracy_score
scores = []
for k in range(1,20):
    knn_model1 = KNeighborsClassifier(n_neighbors=k)
    knn_model1.fit(scaled_X1_train,y1_train)
    pred_test1 = knn_model1.predict(scaled_X1_test)
    scores.append(accuracy_score(y1_test,pred_test1))
scores
```

## We can see accuracy scores are much efficient with scaled data,than unscaled data

```python
# Plotting k values and scores
plt.plot(range(1,20),scores,'--',marker = 'o',markerfacecolor = 'red')
```

### Optimum k Value is 7

```python
# Optimum k value is 7
final_model1 = KNeighborsClassifier(n_neighbors=7,metric = 'euclidean')
final_model1.fit(scaled_X1_train,y1_train)
```

```python
final_train_pred1 = final_model1.predict(scaled_X1_train)
```

```python
#Classification report
print(classification_report(y1_train,final_train_pred1))
```

```python
# Prediction on test data
final_test_pred1 = final_model1.predict(scaled_X1_test)
# Classification report
print(classification_report(y1_test,final_test_pred1))
```

## Observations :
In Recall, Precision, f1-score , accuracy the scaled data has more efficient predictions than with normal data.
