# a. Analyze the  top 5 cities with highest number of female graduates

```python
fem_grad=df.nlargest(5,"female_graduates")
fem_grad
```

# b. Find the 5 cities with lowest number of female graduates and concatenate the two dataframes of top 5 cities with  bottom 5 cities

```python
fem_low=df.nsmallest(5,"female_graduates").sort_values("female_graduates",ascending=False)
fem_low
```

```python
b=pd.concat([fem_grad,fem_low],axis=0)
b
```
