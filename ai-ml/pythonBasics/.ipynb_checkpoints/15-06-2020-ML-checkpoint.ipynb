```python
ndf=df.replace({'temperature':-99999,'windspeed':-99999,'event':'0'},np.nan)
ndf
```

```python
ndf.fillna(0)
```

```python
ndf.fillna(method='ffill')                             # ffill previous value is filled in missing values
```

```python
ndf.isnull().sum()
```

```python
data=ndf.fillna(0)
data
```
