```python
## pairwise plot for numerical values:

ty=tips.columns[tips.dtypes!='category']
val=1
for i in range(len(ty)):
    for j in range(len(ty)):
        plt.subplot(3,3,val)
        plt.scatter(tips[ty[i]],tips[ty[j]])
        plt.title(ty[i]+' vs ' + ty[j])
        val+=1
        
```
