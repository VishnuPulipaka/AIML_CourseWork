# Imputing with missing values:

```python
df.isnull().mean().sort_values(ascending=False)
```

```python
## Cabin --> 77% missing values, make new category as 'New'
dfn.Cabin.fillna('NA',inplace=True)
```

```python
dfn.Cabin.isnull().sum()
```

```python
dfn.Age.describe()
```

```python
# Replacing missing values in Age with median
dfn.Age.fillna(df.Age.median(),inplace=True)
dfn.Age.isnull().sum()
```

```python
## Imputing Embarked column:
dfn.Embarked.fillna(df.Embarked.mode()[0],inplace=True)
dfn.Embarked.isnull().sum()
```

```python
# Imputing the Fare category column
dfn.Fare_Category.isnull().sum()
```

```python
dfn.Fare_Category.fillna(dfn.Fare_Category.mode()[0],inplace=True)
dfn.Fare_Category.isnull().sum()
```

```python
dfn.isnull().mean()
```

```python
# MAKE AGE RANGE
dfn["Age-Range"]=pd.cut(dfn.Age,[0,10,20,30,40,60,80])
dfn[["Age","Age-Range"]].head()
dfn[dfn["Age"]==28]
```

```python
#Encoding of Categorical Variables/Columns
dummy_set=pd.get_dummies(dfn.Fare_Category,drop_first=True)
dummy_set
```

```python
dfn=pd.concat([dfn,dummy_set],axis=1)
```

```python
dfn=pd.concat([dfn,pd.get_dummies(dfn.Cabin,prefix='Cabin',drop_first=True),
               pd.get_dummies(dfn.Embarked,prefix='emb',drop_first=True),
               pd.get_dummies(dfn.Fare_Category,prefix='Fare',drop_first=True),
               pd.get_dummies(dfn['Age-Range'],drop_first=True)],axis=1)
```

```python
dfn
```

```python
# Encoding the sex column
from sklearn.preprocessing import LabelEncoder
dfn['Sex']=LabelEncoder().fit_transform(dfn.Sex)
dfn.head()
```

```python
# Drop the unnecessary columns
dfn.drop(['PassengerId','Name','Age','Ticket','Fare','Embarked','Fare_Category','Cabin','Age-Range']
         ,axis=1,inplace=True)
dfn.shape
```

```python
# Separating the Input and Output Data,dropping the 'Survived' column
X=dfn.drop('Survived',axis=1)
X.shape
```

```python
y=dfn.iloc[:,0]
y.shape
```

## input and output into training data and testing data
###### Training--> on training data
###### Testing --> performance check
###### 891 --> out of 891,75% of 891 rows will be in training ,rest in testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
pd.set_option('display.max_columns',None)
X_test.head(5)
```

# 20-06-2020

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression() #creating an object
```

```python
# we have to apply this object(log_reg) to the training data
log_reg.fit(X_train,y_train) # fit method, we train the model
```

Instead of directly prediciting on test data, checking on traning data itself

```python
#predicting on the training data:
y_train_pred=log_reg.predict(X_train)
```

```python
# comparing y_Train and y_train_red
y_train==y_train_pred
```

```python
## accuracy score -->metric
from sklearn.metrics import accuracy_score
accuracy_score(y_train,y_train_pred)
```
